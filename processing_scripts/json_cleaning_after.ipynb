```python
import pandas as pd

combined_df = []
for df_chunk in pd.read_json('after.json', chunksize = 500000, lines=True):
    trimmed_RT = df_chunk[~df_chunk['text'].str.contains("RT")].copy()
    trimmed_RT['focus'] = trimmed_RT.loc[:,trimmed_RT.columns.str.startswith("liwc.focus")].sum(axis=1)
    trimmed_RT['posneg'] = trimmed_RT.loc[:,trimmed_RT.columns.str.endswith("emo")].sum(axis=1)
   
    moral_value_column = trimmed_RT.columns[trimmed_RT.columns.str.startswith('moral_values')]
    trimmed_RT.drop(moral_value_column, axis=1, inplace=True)
    
    trimmed_RT = trimmed_RT[ (trimmed_RT['focus'] > 0) & (trimmed_RT['posneg'] > 0) ].copy()\
    
    combined_df.append(trimmed_RT)

    print(trimmed_RT.shape)
```

```python
df_new = pd.concat(combined_df, ignore_index=True)

df_new
```

```python
df_new.to_csv('after_cleaned.csv')
```
