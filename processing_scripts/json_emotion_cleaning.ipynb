```python
import pandas as pd
```

```python
df_before_election = pd.read_csv('../data/before_cleaned.csv')
df_after_election = pd.read_csv('../data/after_cleaned.csv')
```

```python
df_after_election.head()
```

```python
def get_highest_emotion(row):
    emotion_dict = {
        'emotion.joy': row['emotion.joy'],
        'emotion.anger': row['emotion.anger'],
        'emotion.sadness': row['emotion.sadness'],
        'emotion.disgust': row['emotion.disgust'],
        'emotion.fear': row['emotion.fear']
    }

    max_emotion = max(emotion_dict, key=emotion_dict.get)
    max_emotion_parsed = max_emotion.replace('emotion.', '')

    return max_emotion_parsed
```

```python
df_before_election['highest_emotion'] = df_before_election.apply(lambda row: get_highest_emotion, axis=1)
df_after_election['highest_emotion'] = df_after_election.apply(lambda row: get_highest_emotion, axis=1)
```

```python
df_before_election.to_csv('../data/before_cleaned_emotion.csv', index=False)
df_after_election.to_csv('../data/after_cleaned_emotion.csv', index=False)
```
